```python
import math
def calculate_c(I):
    c = 0.49239 + I^2 * 0.01792 - I^2 * 7.71e-5 + I^3 * 6.75e-7
    return c

def calculate_I(temps):
    result = 0
    for i in range(0, len(temps)):  # i goes from 1 to 5
        result += math.pow(temps[i] / 5, 1.514)
        print(result)
    return result

# Data
months = ['Gen', 'Feb', 'Mar', 'Apr', 'Mag', 'Giu', 'Lug', 'Ago', 'Set', 'Ott', 'Nov', 'Dic']
values = [0.76, 0.80, 1.02, 1.14, 1.31, 1.33, 1.34, 1.23, 1.05, 0.93, 0.77, 0.72]

# Create DataFrame
a_values = pd.DataFrame({'Month': months, 'Value': values})

# Display the DataFrame
print(a_values)
```
